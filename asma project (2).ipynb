## ADD column to specify Students who are graduated and according to their GPA 

```python

#Students who are graduated and according to their GPA 

conditions = [
    (df['currentStatus']=="graduate")&(df['Cumulative average for the last semester'].between(0.00,2.74, inclusive=True)),
(df['currentStatus']=="graduate")&(df['Cumulative average for the last semester'].between(2.75,3.74, inclusive=True)),
    (df['currentStatus']=="graduate")&(df['Cumulative average for the last semester'].between(3.75,4.49, inclusive=True)),
    (df['currentStatus']=="graduate")&(df['Cumulative average for the last semester'].between(4.50,5.00, inclusive=True))
    ]

# create a list of the values we want to assign for each condition
values = ['Acceptable', 'good', 'very good', 'Excellent']

# create a new column 
df['graduate with rate'] = np.select(conditions, values)

# display updated DataFrame
df.head()
```

## ADD column to specify Students who are graduated and according to how many semesters they took

```python


condition = [
    (df['currentStatus']=="graduate")&(df['StudyDurationSemester'].between(0,10, inclusive=True)),
(df['currentStatus']=="graduate")&(df['StudyDurationSemester'].between(11,15, inclusive=True)),
    (df['currentStatus']=="graduate")&(df['StudyDurationSemester'].between(16,20, inclusive=True))
    ]

# create a list of the values we want to assign for each condition
value = ['regular duration', 'Regular period and half', 'Overstayed regular duration']

# create a new column 
df['graduate with Study Duration'] = np.select(condition, value)

# display updated DataFrame
df.head()
```

## Identify student performance

```python
#identify student performance

condition = [
    (df['currentStatus']=!"graduate"),
    (df['currentStatus']=="graduate")&(df['StudyDurationSemester'].between(16,20, inclusive=False)),
    (df['currentStatus']=="graduate")&(df['StudyDurationSemester'].between(11,20, inclusive=True))&(df['Appreciation']=="good"),
    (df['currentStatus']=="graduate")&(df['Appreciation']=="Acceptable"),
    (df['currentStatus']=="graduate")&(df['StudyDurationSemester'].between(0,10, inclusive=True))&(df['Appreciation']=="good"),
    (df['currentStatus']=="graduate")&(df['Appreciation']=="Excellent"),
    (df['currentStatus']=="graduate")&(df['Appreciation']=="very good")
     
    ]

# create a list of the values we want to assign for each condition
value = ['Low','Low', 'Low','Low','Normal','Normal','Normal']

# create a new column 
df['Performance'] = np.select(condition, value)

# display updated DataFrame
df.head()
```

## Training And Testing dataset

```python
xls = pd.ExcelFile("/Users/hadel/Desktop/Final data  2 class copy.xlsx")
df_test = pd.read_excel(xls, 'Test')
df_train= pd.read_excel(xls, 'Train')
```

```python
import string
from sklearn.datasets import fetch_20newsgroups
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.pipeline import Pipeline
#Train
feature_Train=['Cumulative percentage','GAT','SAAT','011ENG marks','012ENG marks','001MAT marks','012MAT marks','011CSM marks','012CSM  marks']
df_train.columns = df_train.columns.to_series().apply(lambda x: x.strip())

x = df_train[feature_Train] 
y = df_train['Performance']

#Test
feature_Test = ['Cumulative percentage','GAT','SAAT','011ENG marks','012ENG marks','001MAT marks','012MAT marks','011CSM marks','012CSM  marks']
df_test.columns = df_test.columns.to_series().apply(lambda x: x.strip())

xTest = df_test[feature_Test] 
yTest = df_test['Performance']

#varibles
x_train=x
y_train=y
x_test=xTest
y_test=yTest

dt = DecisionTreeClassifier(criterion= 'entropy',max_depth = 10, splitter='best',random_state=50)
dt.fit(x_train, y_train) 
pip_pred = dt.predict(x_test)
print(metrics.classification_report(y_test, pip_pred))
```

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import roc_auc_score

#Train
#feature_Train=['Cumulative percentage','012ENG marks','012CSM  marks','011ENG marks']
feature_Train=['Cumulative percentage','GAT','SAAT','011ENG marks','012ENG marks','001MAT marks','012MAT marks','011CSM marks','012CSM  marks']

df_train.columns = df_train.columns.to_series().apply(lambda x: x.strip())

x = df_train[feature_Train] 
y = df_train['Performance']

#Test
#feature_Test =['Cumulative percentage','012ENG marks','012CSM  marks','011ENG marks']
feature_Test=['Cumulative percentage','GAT','SAAT','011ENG marks','012ENG marks','001MAT marks','012MAT marks','011CSM marks','012CSM  marks']
df_test.columns = df_test.columns.to_series().apply(lambda x: x.strip())

xTest = df_test[feature_Test] 
yTest = df_test['Performance']

#varibles
x_train=x
y_train=y
x_test=xTest
y_test=yTest


rf = RandomForestClassifier(n_estimators=150, criterion="entropy",random_state= 0, max_depth=20, oob_score = True,
                           bootstrap = True)
rf.fit(x_train, y_train)
print('R^2 Training Score: {:.2f} \nOOB Score: {:.2f} \nR^2 Validation Score: {:.2f}'.format(rf.score(x, y), 
                                                                                             rf.oob_score_,
                                                                                             rf.score(x_test, y_test)))
pip_pred = rf.predict(x_test)
#pip_pred = (RF.predict_proba(x_test)[:,1] >= 0.5).astype(bool)
print(metrics.classification_report(y_test, pip_pred))
```
